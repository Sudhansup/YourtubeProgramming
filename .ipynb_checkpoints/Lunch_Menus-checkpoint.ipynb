```python
def create_food(name, isVeg):
    return {'name': name, 'isVeg': isVeg}

def create_veg_food(name):
    return create_food(name,True)

def create_NV_food(name):
    return create_food(name,False)
```

```python
def get_days():
    data = []
    for i in range(7):
        day = datetime.date.today() + datetime.timedelta(days= i)        
        data.append({'Day': (day).strftime('%a'), 'Date': day.strftime('%Y-%m-%d'), 'DayIndex': day.weekday() })
    df_days = pd.DataFrame(data)
    df_days['isVeg'] = df_days['DayIndex'].apply(lambda x: True if x > 0 and x%2 ==1 else False)
    return df_days

get_days()
```

```python
def get_main_foods():
    return pd.DataFrame(map(create_veg_food, [x for x in 'Rice,Roti,BrownRice,Parata'.split(',')]))
get_main_foods()
```

```python
def get_nv_foods():
    return pd.DataFrame(map(create_NV_food, [x for x in 'Biryani, Yakini Pulao'.split(',')]))
get_nv_foods()
```

```python
def get_nv_curries():
    return pd.DataFrame(map(create_NV_food, [x for x in 'Chicken,Egg,Fish,Prawns,Crabs'.split(',')]))
get_nv_curries()
```

```python
def get_veg_curries():
    return pd.DataFrame(map(create_veg_food, [x for x in 'Paneer,Veg/Dalma,Mushroom,AloDum,Cabbage/Cauliflower'.split(',')]))
get_veg_curries()
```

```python
def get_veg_breakfast():
    return pd.DataFrame(map(create_veg_food, [x for x in 'Dosa,Upma,Oats,Veg Sandwich,Sewaya,Fruitbowls,Poha, Maggie/Noodles'.split(',')]))
get_veg_breakfast()
```

```python
def get_nv_breakfast():
    return pd.DataFrame(map(create_NV_food, [x for x in 'Bread_Omlet, Chicken_SandWitch,Boiled_Eggs,Omlets'.split(',')]))
get_nv_breakfast()
```

```python
def get_nv_snacks():
    return pd.DataFrame(map(create_NV_food, [x for x in 'Egg_Chop, Chicken_Kebabs, Chicken_Pakodas, Egg_Chicken_Chowmin'.split(',')]))
get_nv_snacks()
```

```python
def get_veg_snacks():
    return pd.DataFrame(map(create_veg_food, [x for x in 'Upma, Bada,Pakodas,Veg Puffs, Ice_Creams,Veg_Kebabs,Muffins, Pizza'.split(',')]))
get_veg_snacks()
```

```python
def get_all_foods():
    df_main_foods = pd.concat([get_main_foods(),get_nv_foods()])
    df_main_foods.rename(columns={'name':'MainCourse'}, inplace=True)
    
    df_curries = pd.concat([get_nv_curries(),get_veg_curries()]) 
    df_curries.rename(columns={'name':'Curries'}, inplace=True)
    
    df_lunch = df_main_foods.merge(df_curries,on =['isVeg'], how = 'outer')
    df_dinner = df_main_foods.merge(df_curries,on =['isVeg'], how = 'outer')
    
    df_lunch.rename(columns={'MainCourse': 'Lunch_MainCourse', 'Curries': 'Lunch_Curries'}, inplace=True)
    df_dinner.rename(columns={'MainCourse': 'Dinner_MainCourse', 'Curries': 'Dinner_Curries'}, inplace=True)
    
    df_bf = pd.concat([get_veg_breakfast(),get_nv_breakfast()])
    df_snacks= pd.concat([get_veg_snacks(),get_nv_snacks()])
    
    df_bf.rename(columns={'name':'Breakfast'},inplace=True)
    df_snacks.rename(columns={'name':'Snacks'},inplace=True)
    
    df_bf_snacks = df_bf.merge(df_snacks, on = ['isVeg'],how = 'outer') 
      
    df_lunch_dinner = df_lunch.merge(df_dinner, on = ['isVeg'],how = 'outer')  
    
    df_lunch_dinner_bf_snacks = df_lunch_dinner.merge(df_bf_snacks, on = ['isVeg'],how = 'outer')
    
    return df_lunch_dinner_bf_snacks

get_all_foods()
```

```python
def get_menus_for_day():
    df_days = get_days()
    df_foods = get_all_foods()
    
    df_day_foods = df_days.merge(df_foods, on = ['isVeg'], how = 'outer')
    return df_day_foods

get_menus_for_day()
```

```python
def get_final_menu(is_dinner_lunch_same = False, can_eat_veg_on_NV_days = False):
    df_menus = get_menus_for_day()
    #Apply the filter here.
    df_filter_menus = df_menus
    
    return df_filter_menus
get_final_menu()
```
