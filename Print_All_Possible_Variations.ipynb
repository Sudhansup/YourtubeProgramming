```python
#Print all possible variations of the given string
#Input : "cat"
#Output : [cat, cta, act, atc, tca, tac]

def permutation(input_str):
    mainlist = []
    if len(input_str) <= 1:
        return [input_str]
    elif len(input_str) == 2:
        return [input_str, input_str[::-1]]
    else:
        for i in range(len(input_str)):
            for j in permutation(input_str[i+1:] + input_str[:i]):
                mainlist.append(input_str[i] + j)
    return mainlist
```

```python
permutation('stack')
```
